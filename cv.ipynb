```python
import tensorflow as tf
import cv2
import numpy as np

import mtcnn
```

```python
model=tf.keras.models.load_model('mask.h5')


class_Names=['Mask','Wihtout Mask']
face=mtcnn.MTCNN()
```

```python
cap=cv2.VideoCapture(0)
while True:
    ret ,frame=cap.read()
    image=cv2.resize(frame,(224,224))
    face_angel=face.detect_faces(frame)

    image_batch=np.expand_dims(image,axis=0)

    for i in face_angel:
        x,y,w,h=i['box']


        faces=cv2.rectangle(frame,pt1=(x,y),pt2=(x+w,y+h),color=(255,0,0),thickness=2)
        print( class_Names[np.argmax(model.predict(image_batch))])

        



        
    
    

    


    



  
    cv2.imshow('window',frame)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cv2.destroyAllWindows()


```
